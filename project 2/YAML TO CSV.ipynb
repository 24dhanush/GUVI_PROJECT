```python
import os
import yaml
import pandas as pd


# ── 1. Load all YAML file names from folder ──────────────────────────────────
folder_path='/content/drive/My Drive/stock' # ← change this
yaml_files = [
    f for f in os.listdir(folder_path)
    if f.endswith(".yaml") or f.endswith(".yml")
]

print("YAML files found:", yaml_files)
# ── 2. Load each file as a DataFrame ─────────────────────────────────────────
dfs = []

for file_name in yaml_files:
    file_path = os.path.join(folder_path, file_name)

    with open(file_path, "r") as f:
        data = yaml.safe_load(f)

    # yaml.safe_load returns a list of dicts  →  direct DataFrame
    # yaml.safe_load returns a single dict    →  wrap in a list first
    if isinstance(data, list):
        df = pd.DataFrame(data)
    elif isinstance(data, dict):
        df = pd.DataFrame([data])
    else:
        print(f"Skipping {file_name}: unsupported structure ({type(data)})")
        continue

    df["source_file"] = file_name   # optional: track origin
    dfs.append(df)
# ── 3. Concatenate into master DataFrame ─────────────────────────────────────
master_df = pd.concat(dfs, ignore_index=True)

print(master_df.shape)
print(master_df.head(20))
```

```python
import pandas as pd
df= master_df
# Save to CSV
df.to_csv('output.csv', index=False)

print("CSV file saved successfully!")
```

# IMPORT YAMAL FILE AND DOWNLOADE CSV AS INDIVIDUAL **TICKER**

```python
import yaml
import pandas as pd
import os

def yaml_to_csv(input_dir, output_dir):

    os.makedirs(output_dir, exist_ok=True)

    all_data = []

    for filename in os.listdir(input_dir):

        if filename.endswith((".yaml", ".yml")):

            yaml_path = os.path.join(input_dir, filename)

            with open(yaml_path, "r") as f:
                data = yaml.safe_load(f)

            if not isinstance(data, list):
                data = [data]

            all_data.extend(data)

    df = pd.DataFrame(all_data)

    for symbol, group in df.groupby("Ticker"):

        csv_path = os.path.join(output_dir, f"{symbol}.csv")

        group.to_csv(csv_path, index=False)


        print(f"Saved {csv_path}")

yaml_to_csv('/content/drive/My Drive/stock', '/content/drive/My Drive/csv')
```
